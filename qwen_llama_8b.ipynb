# Post Responses Task with Qwen 8B instruct and Llama 8b instruct

### Qwen 8b Instruct

```python
from openai import OpenAI

client = OpenAI(
  base_url="https://openrouter.ai/api/v1",
  api_key="<key>",
)

completion = client.chat.completions.create(
  extra_body={},
  model="qwen/qwen3-vl-8b-instruct",
  messages=[
    {
      "role": "user",
      "content": [
        {
          "type": "text",
          "text": """ You will be given a number of Turkish social media posts.
        Write a response post to each one.
        Return the responses in the format as follows:
        1. [responses 1]
        2. [responses 2]
        .
        
        Here are the text posts to respond to:
        1. her açtığımda belki gülhan şen kaydolmuştur diye search edip havamı aldığım trendy oluşum.
        2. az önce bana allah'ı takip etmem yönünde tavsiye verendir. 'who to follow' maskesinin altında misyonerlik mi yapıyor nedir, anlamış değilim.
        3. (bkz: disney), twitter`ı satın almak için kolları sıvamış durumda...\nmuhtemelen sanal gerçekliği daha ileri taşımak için yapılan bir hamle olsa gerek...
        """
        },

      ]
    }
  ]
)
with open("qwen_8b_post_responses.txt", "w") as f:
    f.write(completion.choices[0].message.content)
```
