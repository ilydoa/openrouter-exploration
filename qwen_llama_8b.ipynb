# Post Responses Task with Qwen 8B instruct and Llama 8b instruct

```python
import os
from dotenv import load_dotenv
import pandas as pd
```

```python
df = pd.read_csv("turkish_text_samples.csv")
text_samples = df["text"].to_list()
```

```python
load_dotenv()
api_key = os.getenv("API_KEY")
```

```python
if api_key is None:
    print("Error: API key not found in environment variables.")
else:
    print("API key loaded successfully.")
```

### Qwen 8b

```python
from openai import OpenAI

client = OpenAI(
  base_url="https://openrouter.ai/api/v1",
  api_key=api_key,
)

completions = []



for sample in text_samples:
  completion = client.chat.completions.create(
    extra_body={},
    model="qwen/qwen3-8b",
    messages=[
      {
        "role": "user",
        "content": [
          {
            "type": "text",
            "text": f""" You will be given a Turkish social media post.
          Write a response post to it.
          
          Here is the text post to respond to:
          {sample}
          """
          },

        ]
      }
    ]
  )

  completions.append(completion.choices[0].message.content)
```

```python
df["qwen_responses"] = completions
```

### Llama 8b Instruct

```python
completions = []

for sample in text_samples:
  completion = client.chat.completions.create(
    extra_body={},
    model="meta-llama/llama-3.1-8b-instruct",
    messages=[
      {
        "role": "user",
        "content": [
          {
            "type": "text",
            "text": f""" You will be given a Turkish social media post.
          Write a response post to it.
          
          Here is the text post to respond to:
          {sample}
          """
          },

        ]
      }
    ]
  )

  completions.append(completion.choices[0].message.content)
```

```python
df["llama_responses"] = completions
```
