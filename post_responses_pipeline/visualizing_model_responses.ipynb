### Qwen and Llama Post Responses

Inspecting the model outputs.

```python
#Necessary imports
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Read data
df = pd.read_csv("data/outputs/post_responses_qwen_llama.csv")
```

```python
df.head()
```

```python
#Plot llama word frequencies
word_freq = df['llama_responses'].str.split(expand=True).stack().value_counts().reset_index()
word_freq.columns = ['word', 'count']
top_words = word_freq.head(20)

plt.figure(figsize=(10, 6))
plt.barh(top_words['word'], top_words['count'])
plt.gca().invert_yaxis() 
plt.title('Top Word Frequencies')
plt.xlabel('Frequency')
plt.ylabel('Words')
plt.show()
```

```python
#Plot qwen word frequencies
word_freq = df['qwen_responses'].str.split(expand=True).stack().value_counts().reset_index()
word_freq.columns = ['word', 'count']
top_words = word_freq.head(20)

plt.figure(figsize=(10, 6))
plt.barh(top_words['word'], top_words['count'])
plt.gca().invert_yaxis() 
plt.title('Top Word Frequencies')
plt.xlabel('Frequency')
plt.ylabel('Words')
plt.show()
```

```python
#Collect response post word countes
df['qwen_word_count'] = df['qwen_responses'].str.split().str.len()
df['llama_word_count'] = df['llama_responses'].str.split().str.len()
```

```python
#Plot qwen word count distributions
plt.figure(figsize=(10, 6))
plt.hist(df['qwen_word_count'], bins=5, color='skyblue', edgecolor='black')
plt.title('Distribution of Word Counts')
plt.xlabel('Word Count')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.7)
plt.show()
```

```python
#Plot Llama word count distributions
plt.figure(figsize=(10, 6))
plt.hist(df['llama_word_count'], bins=5, color='skyblue', edgecolor='black')
plt.title('Distribution of Word Counts')
plt.xlabel('Word Count')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.7)
plt.show()
```
